The purpose of this script is to demonstrate the difference between the Average Treatment Effect (ATE), Average Treatment Effect on the Treated (ATET), and Average Treatment Effect on the Untreated (ATEU). 

In simulated data, I allow for selection into treatment based on the individual treatment effect. Specifically, units that benefit from treatment the most are more likely to select into treatment.

Note that even in a pure randomized experiment, you can still have selection into treatment. This is because customers cannot be forced to take actions, for which you are interested in the impact of. 


**Sound byte** Reminder that you are often estimating the ATT and do not know it. As a refresher:
1. Average Treatment Effect on the Treated (ATT)* among the customers who saw your product, what was the impact on churn?
2. Average Treatment Effect  (ATE)* over all customers, who would churn from seeing your product? 
3. Average Treatment Effect on the Untreated (ATU)* among customers did not see your product, what would the impact on churn have been?

There are a few reasons why ATU is not equal to ATT:
1. Selection into treatment based on individual treatments. This can happen in AB experiments too, depending on what you are studying. For example, some PS4 players will never respond to an XBox coupon. 
2. When you model the control group to look like the treatment group, which is the case for difference-in-differences and OLS. 

I explored (1) in a notebook of simulations here (). I also find the same is true for DML models, even those that estimate individual impacts using doubly robust methods. 



```python
import numpy as np
import pandas as pd
import statsmodels.api as sm


%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression, LinearRegression, Ridge
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import cross_val_predict

import stnomics as st
```

## DGP Functions

```python
def dgp(N=1000, at_mean=1):
    ## Initialize data frame of covariates
    data=pd.DataFrame(
        data={'x0':np.random.uniform(0,1,N)
             ,'x1':np.random.uniform(0,1,N)})
    x_list = [d for d in data.columns if 'x' in d]
        
    ## Determine the baseline outcome
    y0 = np.dot(data[x_list]
                              ,np.random.uniform(1,3,len(x_list)) )
    y0 += np.random.normal(0,0.25, N)
    
    ## Determine a randomly distributed impact
    at = np.random.normal(at_mean,2, N)
    
    ## Determine treatment indicator
    ## based on the baseline outcome and impact and noise
    data['w'] = (np.exp(y0 + at) / (1+np.exp(y0+at)) + np.random.uniform(-1,1,N) > 1 ).astype(float)

    ## Calculate observed outcome
    y_obs = np.dot(data[x_list]
                              ,np.random.uniform(-1,1,len(x_list)) )\
            + np.random.normal(0,0.25,N)
    y_obs += data['w']*at
    data['y'] = y_obs
    data['at'] = at
    data['y0'] = y0
    
    return data
    
```

## Simulate different iterations to show that OLS estimates the ATET


```python
atu_list=[]
att_list=[]
ate_list=[]
ols_est = []
irm_est_ate=[]
irm_est_att=[]
irm_est_atu=[]
sim_times = 200
for s in range(sim_times):
    df= dgp()
    
    ## OLS model
    X = sm.add_constant(df[['x0','x1','w']])
    ols_fit = sm.OLS( df['y'], X).fit()
    ols_estimate = ols_fit.params['w']
    
    ## DML IRM model
    irm_results = st.ate.dml.dml_irm(data_est=df
                      ,split_name='splits',feature_name=['x0','x1']
                      ,outcome_name='y',treatment_name='w'
                      ,ymodel=Ridge(alpha=1)
                      ,tmodel=LogisticRegression(C=1)
                      ,n_data_splits=3
                      ,aux_dictionary={'lower':0.01, 'upper':0.99})    
    df['irm_ite'] = irm_results['treatment_estimate']
    true_atu = df.loc[df['w']==0]['at'].mean()
    true_att = df.loc[df['w']==1]['at'].mean()
    true_ate = df['at'].mean()
    
    # df.groupby('w')['at'].describe()
    atu_list.append(true_atu)
    att_list.append(true_att)
    ate_list.append(true_ate)
    ols_est.append(ols_estimate)
    irm_est_ate.append(df.loc[df['w']==1]['irm_ite'].mean())
    irm_est_att.append(df['irm_ite'].mean())
    irm_est_atu.append(df.loc[df['w']==0]['irm_ite'].mean())
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})

fig,ax=plt.subplots(ncols=3,nrows=2,figsize=(10,8)
                   ,sharex=False, sharey=True)
ax[0,0].scatter(ols_est, ate_list, color='cyan')
ax[0,1].scatter(ols_est, att_list, color='cyan')
ax[0,2].scatter(ols_est, atu_list, color='cyan')
ax[0,0].set_ylabel('True Effect')
ax[0,0].set_title('Avg Effect (ATE)', color='lime')
ax[0,1].set_title('Avg Effect \n Among Treated (ATT)', color='lime')
ax[0,2].set_title('Avg Effect \n Among Untreated (ATU) ', color='lime')

ax[1,0].scatter(irm_est_ate, ate_list, color='royalblue')
ax[1,1].scatter(irm_est_att, att_list, color='royalblue')
ax[1,2].scatter(irm_est_atu, atu_list, color='royalblue')
ax[1,0].set_ylabel('True Effect')
ax[1,0].set_title('Avg Effect (ATE)', color='coral')
ax[1,1].set_title('Avg Effect \n Among Treated (ATT)', color='hotpink')
ax[1,2].set_title('Avg Effect \n Among Untreated (ATU) ', color='hotpink')


for x in[0,1]:
    if x==0:
        labelcolor='lime'
        labelcolorB='lime'
        facecolor='grey'
        model='OLS'
    else:
        labelcolor='hotpink'
        labelcolorB='hotpink'
        facecolor='silver'
        model='DML'
    for r in [0,1,2]:
        ax[x,r].set_xlabel('{0} Estimate'.format(model))
        ax[x,r].set_xlim([1.2,1.6])
        if r==1:
            ax[x,r].grid()
            ax[x,r].xaxis.label.set_color(labelcolorB)
            ax[x,r].yaxis.label.set_color(labelcolorB)
            ax[x,r].tick_params(axis='x', colors=labelcolor)
            ax[x,r].tick_params(axis='y', colors=labelcolor)
            ax[x,r].set_facecolor(facecolor)
        else:
            ax[x,r].grid()
            ax[x,r].xaxis.label.set_color(labelcolor)
            ax[x,r].yaxis.label.set_color(labelcolor)
            ax[x,r].tick_params(axis='x', colors=labelcolor)
            ax[x,r].tick_params(axis='y', colors=labelcolor)
            ax[x,r].set_facecolor(facecolor)
    
fig.tight_layout()
fig.set_facecolor('black')
plt.show()

    
```
