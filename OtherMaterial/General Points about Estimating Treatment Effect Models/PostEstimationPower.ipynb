The purpose of this script is to show that calculating power based on the point estimate and standard error you got from a given dataset is not an accurate representation of power. This is because both the standard error and the estimated difference are noisy. 

You can break this link by calculating the sample size or standard error needed to get the point estimate you got.


## Resources
1. "Post Hoc Power Calculations Are Not Useful", https://library.virginia.edu/data/articles/post-hoc-power-calculations-are-not-useful.
2. Why ex-post power using estimated effect sizes is bad, but an ex-post MDE is not https://blogs.worldbank.org/impactevaluations/why-ex-post-power-using-estimated-effect-sizes-bad-ex-post-mde-not



**sound byte summary:** 

"Our estimate's statistical power is above 80%" should be met with a raised eyebrow and a follow-up. This is because the estimate itself is noisy. An estimate's statistical power could come from an estimate that is large by pure chance, or because your experiment's design has enough statistical power. Statistical power should instead come from the **design.**

I show with simulations that the statistical power based on a given estimate dramatically differs from the **actual statistical power** of the design. By pure chance, you could over-estimate the true difference and similarly over-estimate your power. 

So how do you create trust in your results? You can infer the minimum detectable effect (MDE) based on the data. This removes the noise that over-estimates your power, but still retains some noise from the data. 

This post was inspired by these two blog posts, which covers this in more detail: https://library.virginia.edu/data/articles/post-hoc-power-calculations-are-not-useful and https://blogs.worldbank.org/impactevaluations/why-ex-post-power-using-estimated-effect-sizes-bad-ex-post-mde-not.


## DGP and Calculation Functions


```python
from statsmodels.stats.power import ttest_power, tt_ind_solve_power
from statsmodels.stats.weightstats import ttest_ind

'''
statsmodels.stats.power.tt_ind_solve_power(effect_size=None
, nobs1=None
, alpha=None
, power=None
, ratio=1.0
, alternative='two-sided')Â¶
'''
import pandas as pd
import numpy as np

%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
```

```python
sd = 1
def dgp(N0=100,N1=100
       , effect = 0.02):
    ## Generate a dataset
    control = np.random.normal(0,sd,N0)
    treat = np.random.normal(0,sd,N1)
    ## effect can be interpreted as standard deviation difference here
    treat += np.abs(treat)*effect
    
    se = sd / np.sqrt(N0+N1)
    return control, treat
    
    
```

```python
true_treatment_effect = 0.5
N0 = 10
N1 = 50

true_se = sd / np.sqrt(N0+N1)
```

```python
tt_ind_solve_power(effect_size=true_treatment_effect/true_se
                                    , nobs1=N0
        , alpha=0.05
        , power=None
        , ratio =N1/N0
        , alternative='two-sided')
```

```python

df_sim = pd.DataFrame()

ex_ante_power = tt_ind_solve_power(effect_size=true_treatment_effect
                                    , nobs1=N0
        , alpha=0.05
        , power=None
        , ratio =N1/N0
        , alternative='two-sided')
true_mde = tt_ind_solve_power(effect_size=None
                                    , nobs1=N0
        , alpha=0.05
        , power=0.80
        , ratio =N1/N0
        , alternative='two-sided')

for r in range(250):
    ## Generate df_test
    df_test = dgp(N0=N0, N1=N1, effect=true_treatment_effect)
    ## Do Analysis
    tresult = ttest_ind(df_test[0],df_test[1])
    diff = np.mean(df_test[1]) - np.mean(df_test[0])
    se = np.abs(diff / tresult[0]) 
    effect_size_calc = diff/se
    ## Calculcate power given Analysis results
    post_power = tt_ind_solve_power(effect_size=effect_size_calc
                                    , nobs1=len(df_test[0])
        , alpha=0.05
        , power=None
        , ratio =len(df_test[1])/len(df_test[0])
        , alternative='two-sided')
    
    ## Instead, calculate how much more treated units needed
    ## to detect the expected effect size    
    try:
        post_ratio = tt_ind_solve_power(effect_size=effect_size_calc
                                        , nobs1=N0
            , alpha=0.05
            , power=0.80
            , ratio = None
            , alternative='two-sided')
        n1_calc = post_ratio*len(df_test[0])
        
        mde = tt_ind_solve_power(effect_size=None
                                        , nobs1=N0
            , alpha=0.05
            , power=0.80
            , ratio =len(df_test[1])/len(df_test[0])
            , alternative='two-sided')*se
        
    except:
        post_ratio=np.nan
        n1_calc=np.nan
    
    df_sim = pd.concat([df_sim, pd.DataFrame(index=[r], data={'diff':diff, 'se':se, 'pvalue':tresult[1]
                                                             ,'post_power':post_power
                                                             ,'true_power':ex_ante_power
                                                             , 'mde':mde
                                                             ,'Treated_Needed':n1_calc
                                                             ,'Incremental Treated_Needed':n1_calc-N1})])
```

```python
print('Among 95% stat sig estimates:')
display(df_sim.loc[df_sim['pvalue']<0.05][['post_power','true_power']].mean() )
print('Among 95% not stat sig estimates:')
display(df_sim.loc[df_sim['pvalue']>0.05][['post_power','true_power']].mean() )
```

```python
fig,ax = plt.subplots(ncols=3, nrows=1, figsize=(10, 4 ))
ax[0].scatter(x=df_sim['diff'],y= df_sim['post_power'], s=2)
ax[0].plot([true_treatment_effect]*2, [0,1], label='True Difference', color='black', linestyle='--')
ax[0].plot( [df_sim['diff'].min(), df_sim['diff'].max()], [ex_ante_power]*2, label='Ex-Ante Power', color='magenta', linestyle='dotted')
ax[0].legend(loc='center right')
ax[0].set_xlabel('Estimated Difference')
ax[0].set_ylabel('Ex-Post Power:\n  Power given Estimated Difference')
ax[0].grid()


ax[1].scatter(df_sim['diff'], df_sim['mde'], color='goldenrod', alpha=0.45)
ax[1].plot([true_treatment_effect]*2, [df_sim['mde'].min(),df_sim['mde'].max()], label='True Difference', color='black', linestyle='--')
ax[1].set_xlabel('Estimated Difference')
ax[1].set_ylabel('MDE with 80% Power')
ax[1].grid()

ax[2].scatter(df_sim['diff'], df_sim['pvalue'], color='teal', alpha=0.45)
ax[2].set_xlabel('Estimated Difference')
ax[2].set_ylabel('P-value')
ax[2].grid()


fig.set_facecolor('lightgreen')
fig.tight_layout()
plt.show()
```
